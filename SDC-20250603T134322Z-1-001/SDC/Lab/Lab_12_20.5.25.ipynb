```python
# Viết chương trình tính chỉ số BMI với Chiều cao, cân nặng đã biết
# (dùng OOP)
class BMI:
  def __init__(self,height, weight):
    self.__height = height
    self.__weight = weight
  @property
  def height(self):
    return self.__height
  @height.setter
  def height(self,height):
    self.__height = height
  @property
  def weight(self):
    return self.__weight
  @weight.setter
  def weight(self,weight):
    self.__weight = weight
  def tinhBMI(self):
    bmi = self.__weight / (self.__height * self.__height)
    if bmi < 18.5:
      result = "Gầy"
    elif bmi <23:
      result = "Bình Thường"
    elif bmi <25:
      result = "Thừa cân"
    elif bmi <30:
      result = "Béo phì độ I"
    elif bmi <40:
      result = "Béo phì độ II"
    else :
      result = "Béo phì độ III"
    return result
```

```python
o1 = BMI(1.75,78)
o1.tinhBMI()
```
